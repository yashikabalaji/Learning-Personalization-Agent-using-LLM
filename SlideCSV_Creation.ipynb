```python
import PyPDF2
import csv

def pdf_to_csv(pdf_path, csv_path):
    try:
        with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages)

            with open(csv_path, 'w', encoding='utf-8', newline='') as csv_file:
                csv_writer = csv.writer(csv_file)
                csv_writer.writerow(['Slide Number', 'Content'])

                for page_num in range(num_pages):
                    page = pdf_reader.pages[page_num]
                    content = page.extract_text()
                    csv_writer.writerow([page_num + 1, content])

        print(f'Successfully exported PDF content to CSV. CSV saved to: {csv_path}')

    except Exception as e:
        print(f'Error: {e}')

# Example usage
pdf_file_path = 'dbs-8.pdf'
csv_file_path = 'slide_content_8.csv'
pdf_to_csv(pdf_file_path, csv_file_path)
```
